# Exploratory Data Analysis (EDA)

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import os

# Configure matplotlib for inline plotting
%matplotlib inline
```

```python
# Load environment variables from .env file
load_dotenv()

# Get directory paths from environment variables
NOTEBOOKS_DIR_PATH = os.getenv('NOTEBOOKS_DIR_PATH')
DATASETS_DIR_PATH = os.getenv('DATASETS_DIR_PATH')
REPORTS_DIR_PATH = os.getenv('REPORTS_DIR_PATH')

# Define file paths
processed_data_path = os.path.join('..', DATASETS_DIR_PATH, 'processed_data.csv')

# Load the processed data
df = pd.read_csv(processed_data_path)
```

## Data Inspection

Let's start by inspecting the dataset to understand its structure and content.


```python
# Display the first few rows of the dataset
df.head()
```

```python
# Get dataset shape
print(f"Dataset Shape: {df.shape}")
```

```python
# Check for missing values
print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Get basic statistics
df.describe()
```

## Correlation Analysis

We will analyze the correlation between different features and the target variable to identify significant relationships.


```python
# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

### Features Correlating with Student Performance

Let's visualize the correlation of features specifically with the target variable `passed`.

```python
# Correlation with the target variable
corr_with_target = df.corr()['passed'].sort_values(ascending=False)
print("Correlation with Target Variable 'passed':")
print(corr_with_target)
```

```python
# Plotting the correlations
plt.figure(figsize=(6, 10))
sns.heatmap(corr_with_target.to_frame(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Features Correlating with 'passed'", fontsize=14)
plt.show()
```

**Interpretation:**

- **Positive Correlations:** Features like `higher`, `Medu`, `Fedu`, and `studytime` show a positive correlation with passing the exam.
- **Negative Correlations:** Features like `failures`, `goout`, and `absences` show a negative correlation with passing the exam.

These features are significant and will be explored further in the following sections.


## Data Visualization

We will create visualizations to understand the distribution and relationships of key features.


### Distribution of Target Variable


```python
# Countplot of the target variable
sns.countplot(x='passed', data=df)
plt.title('Distribution of Target Variable: Passed', fontsize=14)
plt.xlabel('Passed (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

**Observation:**

This shows the number of students who passed versus those who did not.


### Effect of Study Time on Passing Rate

```python
# Bar plot of study time vs. passing rate
sns.barplot(x='studytime', y='passed', data=df)
plt.title('Study Time vs. Passing Rate', fontsize=14)
plt.xlabel('Study Time (Normalized)', fontsize=12)
plt.ylabel('Passing Rate', fontsize=12)
plt.show()
```

**Observation:**

Students who spend more time studying have a higher passing rate.


### Impact of Failures on Passing Rate


```python
# Bar plot of failures vs. passing rate
sns.barplot(x='failures', y='passed', data=df)
plt.title('Number of Failures vs. Passing Rate', fontsize=14)
plt.xlabel('Number of Past Class Failures (Normalized)', fontsize=12)
plt.ylabel('Passing Rate', fontsize=12)
plt.show()
```

**Observation:**

An increase in past failures correlates with a lower passing rate.

### Relationship between Parental Education and Student Success

```python
# Mother's education vs. passing rate
sns.barplot(x='Medu', y='passed', data=df)
plt.title("Mother's Education vs. Passing Rate", fontsize=14)
plt.xlabel("Mother's Education Level (Normalized)", fontsize=12)
plt.ylabel('Passing Rate', fontsize=12)
plt.show()
```

```python
# Father's education vs. passing rate
sns.barplot(x='Fedu', y='passed', data=df)
plt.title("Father's Education vs. Passing Rate", fontsize=14)
plt.xlabel("Father's Education Level (Normalized)", fontsize=12)
plt.ylabel('Passing Rate', fontsize=12)
plt.show()
```

**Observation:**

Higher parental education levels are associated with higher student passing rates.

### Effect of Going Out Frequency on Passing Rate

```python
# Bar plot of goout vs. passing rate
sns.barplot(x='goout', y='passed', data=df)
plt.title('Going Out Frequency vs. Passing Rate', fontsize=14)
plt.xlabel('Going Out Frequency (Normalized)', fontsize=12)
plt.ylabel('Passing Rate', fontsize=12)
plt.show()
```

**Observation:**

Students who frequently go out with friends tend to have lower passing rates.

## General Insights

Based on the visualizations and correlation analysis:

- **Positive Influencers:**
  - **Higher Education Aspiration (`higher`):** Students desiring higher education are more likely to pass.
  - **Parental Education (`Medu`, `Fedu`):** Higher parental education correlates with student success.
  - **Study Time (`studytime`):** More study time increases passing rates.

- **Negative Influencers:**
  - **Past Failures (`failures`):** More past failures decrease the likelihood of passing.
  - **Going Out (`goout`):** Frequent social outings correlate with lower passing rates.
  - **Absences (`absences`):** Higher absences are associated with lower passing rates.

## Recommendations

- **Encourage Study Habits:** Promote effective study strategies to increase study time.
- **Parental Involvement:** Engage parents in supporting their child's education.
- **Monitor Attendance:** Address issues leading to high absences.
- **Balance Social Activities:** Encourage a balance between social life and academics.
